```python
import numpy as np
#criaçao de um dataframe de teste! Se atentar a escrita
df_salarios = pd.DataFrame({
    'nome': ["Ana", "Bruno", "Carlos", "Danielle", "Val"],
    'salarios': [4000, np.nan, 5000, np.nan, 10000]
})
#calcula a medias dos valores(e arredonda) e substitui os nulos pelo resultado
df_salarios['salarios_media'] = df_salarios['salarios'].fillna(df_salarios['salarios'].mean().round(2))
df_salarios
'''calcula a mediana dos valores e substitui os nulos pelo resultado'''
df_salarios['salarios_media'] = df_salarios['salarios'].fillna(df_salarios['salarios'].median())

df_salarios
```

```python
df_temperatura = pd.DataFrame ({
    "Dia": ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"],
    "Temperatura": [30, np.nan, np.nan, 28, 27],

})
df_temperatura["preenchido_ffill"] = df_temperatura["Temperatura"].fillna(method='ffill')
df_temperatura
```

```python
df_cidades = pd.DataFrame({
    "nome": ["Ana", "Bruno", "Carlos", "Danielle", "Val"],
    "cidade": ["sao paulo", np.nan, "Curitiba", np.nan, "Belem"]

})
df_cidades["cidade_preenchida"] = df_cidades["cidade"].fillna("Nao informado")
display(df_cidades)
```
