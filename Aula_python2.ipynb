```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")
```

```python
linhas, colunas = df.shape[0], df.shape[1]
print("linhas: ", linhas)
print("colunas: ", colunas)
```

```python
df.rename(columns={
    'ano': 'ano',
    'nivel_experiencia': 'senioridade',
    'tipo_emprego': 'contrato',
    'cargo': 'cargo',
    'salario': 'salario',
    'moeda_salario': 'moeda',
    'salario_em_usd': 'usd',
    'residencial_empregado': 'residencia',
    'taxa_remoto': 'remoto',
    'localizacao_empresa': 'empresa',
    'tamanho_empresa': 'tamanho_empresa'
}, inplace=True)

display(df.columns)
```

```python
df["senioridade"].value_counts()
```

```python
df["contrato"].value_counts()
```

```python
df["remoto"].value_counts()
```

```python
df["tamanho_empresa"].value_counts()
```

```python
senioridade_mapping = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}
df['senioridade_pt'] = df['senioridade'].map(senioridade_mapping)
display(df['senioridade_pt'].value_counts())
```

```python
contrato_mapping = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
}
df['contrato_pt'] = df['contrato'].map(contrato_mapping)
display(df['contrato_pt'].value_counts())
```

```python
tamanho_empresa_mapping = {
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
}
df['tamanho_empresa_pt'] = df['tamanho_empresa'].map(tamanho_empresa_mapping)
display(df['tamanho_empresa_pt'].value_counts())
```

```python
display(df.columns)
remoto_mapping = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}
df['remoto_pt'] = df['remoto'].map(remoto_mapping)
display(df['remoto_pt'].value_counts())
```
